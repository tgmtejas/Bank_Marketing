```python
fig, ax = plt.subplots(figsize=(6, 6))
plt.rcParams["figure.figsize"] = (15,10)

sns.distplot(Bank['age'], hist=True, kde=False, 
             bins=10, color = 'blue',
             hist_kws={'edgecolor':'black'})
plt.title('Age')
```

```python
fig, ax = plt.subplots(figsize=(20, 6))

sns.countplot('campaign', hue = 'y',data=Bank, order=sorted(Bank['campaign'].value_counts().index))
plt.title("Number of contacts during campaign")
```

```python
fig, ax = plt.subplots(figsize=(16, 6))

sns.countplot('job', hue = 'y',data=Bank, order=Bank['job'].value_counts().index)
plt.title("Job Status")
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

sns.countplot('marital', hue = 'y',data=Bank, order=Bank['marital'].value_counts().index)
plt.title("Marital Status")
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

sns.countplot('education', data=Bank, hue = 'y',order=Bank['education'].value_counts().index, palette="Set2")
plt.title(" Education Status")
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

sns.countplot('housing', data=Bank, hue = 'y',order=Bank['housing'].value_counts().index, palette = "Blues_d")
plt.title(" Housing loan respondents")
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

sns.countplot('loan', data=Bank, hue = 'y',order=Bank['housing'].value_counts().index, palette = "Blues_d")
plt.title(" Personal loan respondents")
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

sns.countplot('contact', data=Bank, hue = 'y',order=Bank['contact'].value_counts().index, palette = "dark")
plt.title("Contact Type")
```

```python
fig, ax = plt.subplots(figsize=(16, 6))

sns.countplot('day', data=Bank, order=sorted(Bank['day'].value_counts().index), palette = "bright")
plt.title("Counts by Day of Month")
```

```python
fig, ax = plt.subplots(figsize=(16, 6))
months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
sns.countplot('month', data=Bank, order=months, palette = "bright")
plt.title("Counts by Month")
```

```python
fig, ax = plt.subplots(figsize=(16, 6))

sns.countplot('poutcome', data=Bank, hue = 'y',order=Bank['poutcome'].value_counts().index, palette = "bright")
plt.title("Number of respondents by the outcome of the previous campaign")
```
